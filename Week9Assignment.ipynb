```python
#3.Check whether all the zones generate the same amount of sales.

#Null hypothesis: All zones generate the same amount of sales.

#Alternate Hypothesis: All zones does not generate the same amount of sales.

#Significance value: 0.05
```

```python
data.info()
```

```python
fvalue,pvalue=stats.f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
fcrit=stats.f.ppf(1-0.05,28,112) #n-1=28,n-k=116-4=112
```

```python
print(fvalue,pvalue,fcrit)
```

```python
if fvalue>fcrit and pvalue<0.05:
   print('Reject Null Hypothesis: All Zones generate same amount of sales')
else:
   print('Reject Alternate Hypothesis: All zones generate different amount of sales')
```

```python
#Since pvalue is less than significance level, Null Hypothesis is rejected. Hence it can be stated that all zones generate different amount of sales!
```
